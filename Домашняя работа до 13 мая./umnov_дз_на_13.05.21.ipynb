```
df = pd.read_csv("train (7).csv")
df
```

```
df['price_range'].value_counts()
```

```
X = df.drop(['price_range'], axis=1)
y = df['price_range']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```
#логистическая регрессия 
model = LogisticRegression()

model.fit(X_train, y_train)
answer_pred = model.predict(X_train)
print(model.score(X_test, y_test))
```

```
#XGBoost
modelXGBoost = XGBClassifier()
modelXGBoost.fit(X_train, y_train)

print(modelXGBoost.score(X_test, y_test))
```

```
X = df.iloc[:, :20]
Y = df.iloc[:, 20]

modeln = Sequential()
modeln.add(Dense(1, input_dim = 20, activation = 'relu'))
modeln.add(Dense(3, activation = 'relu'))
modeln.add(Dense(1, activation = 'sigmoid'))
modeln.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ['accuracy', 'Recall'])
modeln.fit(X, Y, epochs = 5, batch_size = 10)

y_pred = modeln

# 5 эпох добавил сам для доп. проверки
```
