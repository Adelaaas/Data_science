```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

## Загрузка данных

```python
data = pd.read_csv("country_vaccinations.csv")
```

```python
data
```

```python
data = data.drop(columns=['source_name','source_website','date'], axis =1)
data = pd.get_dummies(data, columns=['country', 'iso_code'])
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

data['vaccines'] = le.fit_transform(data['vaccines'])
data
```

```python
data.iloc[0,1]
```

```python
train, test = train_test_split(data, shuffle=True, 
                                         test_size=.3, 
                                         random_state=42)
```

```python
test
```

## Исследование данных

```python
# размерность у train - это датафрейм в котором набор данных для обучения
print(train.shape)
# размерность у test
# shape - возвращает размерность датафрейма (кол-во строк, кол-во столбцов)
print(test.shape)
#какие типы данных представлены в наборе
# info()
# dtypes
print(train.info())
print(test.dtypes)

#есть ли пустые значения в данных? Nan
# name_dataframe.isna() - проверку на пустые значение
print("________________________________________")
print(train.isna().sum())
print("________________________________________")
print(test.isna().sum())
```

## Подготовка данных

```python
# обработаем сначала train

train = train.dropna(axis = 0)

print(train.isna().sum())
```

```python


test = test.dropna(axis = 0)

print(test.isna().sum())
```

## Обучение модели

```python
X_tr = train.drop(['vaccines'], axis=1)
y_tr = train['vaccines']

model = LogisticRegression()
# модель учится, для этого ей нужны правильные
model.fit(X_tr, y_tr)

X_te = test.drop(['vaccines'], axis=1)
y_te = test['vaccines']

# тестируем, для признаков у которых нет ответов, мы предсказывает ответ
answers_pred = model.predict(X_te)
# answers_pred - ответы которые вернула модель для X_test
# y_test - это правильные ответы для X_test
print(model.score(X_te, y_te))
```
