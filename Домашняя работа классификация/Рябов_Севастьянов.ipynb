```python
df = df.dropna()
```

```python
df = pd.get_dummies(df, columns=['_unit_state', 'profile_yn'])
df
```

```python
df[df['gender'] == 'brand'] = None
df[df['gender'] == 'unknown'] = None
df = df.dropna()
df
```

####нужно ещё преобразовать male female в 0 1

```python
# Используем LabelEncoder, тк это будет целевой переменной
from sklearn import preprocessing as pre

l = pre.LabelEncoder()
mapped = l.fit_transform(df['gender'])
df['gender'] = mapped
df
```

```python
#Делаем масштабирование признаков: нормализацию и стандартизацию
X = df.drop(['gender'], axis = 1)
y = df['gender']
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_st = scaler.fit_transform(X)
print(X_st)
print('###################'*5)
scaler = MinMaxScaler()
X_norm = scaler.fit_transform(X)
```

```python
#svm с обычными данными + кросвалидация + оценка
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.svm import SVC
model=SVC()
cvs0 = cross_val_score(estimator=model, X=X, y=y, cv=6, scoring='accuracy')
cvs1 = cross_val_score(estimator=model, X=X, y=y, cv=6, scoring='recall')
cvs2 = cross_val_score(estimator=model, X=X, y=y, cv=6, scoring='precision')
print("accuracy")
print(cvs0.mean())
print("recall")
print(cvs1.mean())
print("precision")
print(cvs2.mean())
```

```python
#svm с нормализированными данными + кросвалидация + оценка
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.svm import SVC
model=SVC()
cvs0 = cross_val_score(estimator=model, X=X_norm, y=y, cv=6, scoring='accuracy')
cvs1 = cross_val_score(estimator=model, X=X_norm, y=y, cv=6, scoring='recall')
cvs2 = cross_val_score(estimator=model, X=X_norm, y=y, cv=6, scoring='precision')
print("accuracy")
print(cvs0.mean())
print("recall")
print(cvs1.mean())
print("precision")
print(cvs2.mean())
```

```python
#svm с стандартизированными данными + кросвалидация + оценка
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.svm import SVC
model=SVC()
cvs0 = cross_val_score(estimator=model, X=X_st, y=y, cv=6, scoring='accuracy')
cvs1 = cross_val_score(estimator=model, X=X_st, y=y, cv=6, scoring='recall')
cvs2 = cross_val_score(estimator=model, X=X_st, y=y, cv=6, scoring='precision')
print("accuracy")
print(cvs0.mean())
print("recall")
print(cvs1.mean())
print("precision")
print(cvs2.mean())
```

```python
#knn с обычными данными
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=1/3, random_state=42)
knn = KNeighborsClassifier(5)

knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
confusion_matrix(y_test,y_pred)

print("Accuracy:", accuracy_score(y_test,y_pred))
print("Recall:", recall_score(y_test,y_pred))
print("Precision:", precision_score(y_test,y_pred))
```

```python
#крос валидация
from sklearn.model_selection import cross_val_score
cvs = cross_val_score(knn, X, y, cv=10)
print(cvs)
print(cvs.mean())
```
