нейронка

```python
X = df.iloc[:,0:12]  # y = ax [диапазон строк, диапазон столбцов]
Y = df.iloc[:,20] 
X
```

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
#
model = Sequential()
#Входной слой с 12ю признаками, потом 4 скрытых слоя, потом выходной слой
model.add(Dense(12, input_dim=12, activation='relu'))

model.add(Dense(10, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(3, activation='relu'))

model.add(Dense(1, activation='softmax'))

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy', 'Recall'])
model.fit(X, Y, epochs = 15)
```

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
#
model = Sequential()
#Входной слой с 12ю признаками, потом 4 скрытых слоя, потом выходной слой
model.add(Dense(12, input_dim=12, activation='tanh'))

model.add(Dense(10, activation='tanh'))
model.add(Dense(8, activation='tanh'))
model.add(Dense(5, activation='tanh'))
model.add(Dense(3, activation='tanh'))

model.add(Dense(1, activation='softmax'))

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy', 'Recall'])
model.fit(X, Y, epochs = 15)
```

Видно, что всем известный и проверенный временем метод опорных векторов тут справляется намного лучше чем моя самодельная нейронка, с ненастроенными гиперпараметрами.
