```python
#удаляем ненужные столбцы
APStest = APStest.drop(['Unnamed: 0', 'id'], axis=1)
APStrain = APStrain.drop(['Unnamed: 0', 'id'], axis=1)
```

```python
#разбиваем категориальные признаки с маеьким количеством уникальных значений на разные столбцы
APStest = pd.get_dummies(APStest, columns=['Gender','Customer Type','Type of Travel','Class','satisfaction'])
APStrain = pd.get_dummies(APStrain, columns=['Gender','Customer Type','Type of Travel','Class','satisfaction'])
```

```python
#убираем пустые значения
APStest=APStest.dropna()
```

```python
#убираем пустые значения
APStrain=APStrain.dropna()
```

```python
APStrain
```

```python
#удаляем ненужные столбцы (снова)
X_train = APStrain.drop(['satisfaction_neutral or dissatisfied', 'satisfaction_satisfied'], axis = 1)
y_train = APStrain['satisfaction_satisfied']
X_test = APStest.drop(['satisfaction_neutral or dissatisfied', 'satisfaction_satisfied'], axis = 1)
y_test = APStest['satisfaction_satisfied']

print(X_test)
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_st_train = scaler.fit_transform(X_train[:7000])
X_st_test = scaler.fit_transform(X_test[:7000])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_norm_train = scaler.fit_transform(X_train[:7000])
X_norm_test = scaler.fit_transform(X_test[:7000])
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix

scores = []
for i in range(1,15):
    knn = KNeighborsClassifier(i)
    knn.fit(X_st_train,y_train[:7000])
    answers = knn.predict(X_st_test)
    
    scores.append(knn.score(X_st_test, y_test[:7000]))

print(scores) #[0.8841428571428571, 0.882, 0.9001428571428571, 0.8917142857142857, 0.9035714285714286, 0.8992857142857142, 0.9047142857142857, 0.9017142857142857, 0.9044285714285715, 0.902, 0.9042857142857142, 0.9002857142857142, 0.9038571428571428, 0.9007142857142857]
# Значит 7 соседей является наилучшим выбором.
```

```python
max_score = max(scores)
scores_ind = [i for i, v in enumerate(scores) if v == max_score]
knn = KNeighborsClassifier(7)
knn.fit(X_st_train,y_train[:7000])
y_pred = knn.predict(X_st_test)
confusion_matrix(y_test[:7000],y_pred)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score

y_test = y_test[:7000]
y_train = y_train[:7000]

print("Accuracy:", accuracy_score(y_test,y_pred))
print("Recall:", recall_score(y_test,y_pred))
print("Precision:", precision_score(y_test,y_pred))

# Accuracy: 0.9047142857142857
# Recall: 0.83338767525269
# Precision: 0.9424778761061947


# Все показатели довольно высокие.
```

```python
from sklearn.svm import SVC

svm = SVC()
svm.fit(X_st_train, y_train)
y_pred_svm = svm.predict(X_st_test)
```

```python
print(confusion_matrix(y_test,y_pred_svm))
print("Accuracy:", accuracy_score(y_test,y_pred_svm))
print("Recall:", recall_score(y_test,y_pred_svm))
print("Precision:", precision_score(y_test,y_pred_svm))

# Accuracy: 0.9325714285714286
# Recall: 0.9041408542549723
# Precision: 0.9396814639105388

#Метод опорных векторов работает для этой задачи лучше, чем KNN почти по всем параметрам. Разве что Precision стал незначительно меньше.
```
