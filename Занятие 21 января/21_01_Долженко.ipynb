```
df = pd.read_csv("country_vaccinations.csv")
df
```

```
# ИССЛЕДОВАНИЕ ДАННЫХ
print(df.shape, "\n________________________________\n")
print(df.info(), "\n________________________________\n")
print(df.isna().sum(), "\n________________________________\n")
```

```
# ОЧИСТКА ДАННЫХ
df = df.drop(['people_fully_vaccinated', 'daily_vaccinations_raw', 'people_fully_vaccinated_per_hundred'], axis = 1)
df
```

```
df['iso_code'] = df['iso_code'].fillna(df['iso_code'].mean)
df['total_vaccinations'] = df['total_vaccinations'].fillna(df['total_vaccinations'].mean)
df['people_vaccinated'] = df['people_vaccinated'].fillna(df['people_vaccinated'].mean)
df['daily_vaccinations'] = df['daily_vaccinations'].fillna(df['daily_vaccinations'].mean)
df['total_vaccinations_per_hundred'] = df['total_vaccinations_per_hundred'].fillna(df['total_vaccinations_per_hundred'].mean)
df['people_vaccinated_per_hundred'] = df['people_vaccinated_per_hundred'].fillna(df['people_vaccinated_per_hundred'].mean)
df['daily_vaccinations_per_million'] = df['daily_vaccinations_per_million'].fillna(df['daily_vaccinations_per_million'].mean)

df = df.dropna(how='any')

print(df.shape, "\n________________________________\n")
print("___________После обработки всех пустых значений______________")
print(df.isna().sum())
```

```
# сколько каждой вакцины в наборе
print(df['vaccines'].value_counts())

# сколько источников вакцинирования в наборе? сколько записей соответствует каждому источнику?
print("\n______кол-во источников_______\n", len(df['source_name'].unique()))
print('\n________кол-во записей каждого источника________\n', df['source_name'].value_counts())
```
