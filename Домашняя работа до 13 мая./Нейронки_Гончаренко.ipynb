```python
# Классификация - система группировки объектов исследования или наблюдения в соответствии с их общими признаками.
# Ансамблевые методы - это парадигма машинного обучения, в которой несколько моделей обучаются для решения одной проблемы и объединяются для повышения производительности.
# Нейронная сеть - при помощи искусственных нейронов моделирует работу человеческого мозга (нейронов), решающего определенную задачу, самообучается с учетом предыдущего опыта. И с каждым разом совершает все меньше ошибок.
# Метрики - Accuracy, Precision, Recall, F1-Score, score
```

```python
#SVC
X = df.drop(['price_range'], axis = 1)
y = df['price_range']
#разделяем данные на тестовые-тренировочные
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=42)

accuracy_score = []
#создаём модель и выводим результат
for i in range(1,20):
    svc = SVC(i)
    svc.fit(X_train,y_train)
    answers = svc.predict(X_test) # метод predict возвращает вектор ответов для X_test
    accuracy_score.append(svc.score(X_test, y_test))
print(max(accuracy_score))
```

```python
#XGBoost
#создаём модель
modelXB = XGBClassifier()
modelXB.fit(X_train, y_train)
#выводим результат
print(modelXB.score(X_test, y_test))
```

```python
# Нейронка
X = df.iloc[:, :19]
Y = df.iloc[:, 19]
# создаём модель
modelN = Sequential()
modelN.add(Dense(1, input_dim = 19, activation = 'relu'))
modelN.add(Dense(5, activation = 'relu'))
modelN.add(Dense(1, activation = 'sigmoid'))
modelN.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ['accuracy', 'Recall'])
modelN.fit(X, Y, epochs = 100, batch_size = 10)

y_pred = modelN
```
