```python
import pandas as pd


df = pd.read_csv("C:\\Users\In9.var\Desktop\Projects\data\\train (7).csv") # загружаем датасет
df
```

```python
df.groupby('price_range').count() # число каждого класса. прекрасные данные. спасибо!
```

```python
X = df.drop(['price_range'], axis=1)
y = df['price_range']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier as knn

model1 = knn()
model1.classes_ = 4
model1.fit(X_train, y_train)
model1.score(X_test, y_test) # создаю модель knn, делаю кол-во классов 4, обучаю, тестирую. Совсем неплохой результат
```

```python
from catboost import CatBoostClassifier as cbc, Pool

scores = []

for it in range(1, 5):
    for de in range(1, 5):
        for lr in range(1, 11):
            model2 = cbc(iterations=it, depth=de, learning_rate=lr/10, loss_function='MultiClass', verbose=True, classes_count=4)
            model2.fit(X_train, y_train)
            scores.append(model2.score(X_test, y_test))
max(scores) # Создаю catboost для классификации, перебираем циклом параметры модели, выбираем лучшее из получившихся
```

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy

numpy.random.seed(42)
```

```python
X = df.iloc[:,0:6]
y = df.iloc[:,20]
scores=[]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)

for i in range(4, 15):
    for j in range(4, 10):
        for k in range(4, 10):
            for l in range(4, 10):
                model3 = Sequential([
                    Dense(i, input_dim=6, activation='sigmoid'),
                    Dense(j, activation='sigmoid'),
                    Dense(k, activation='sigmoid'),
                    Dense(l, activation='sigmoid'),
                    Dense(1, activation='softmax')
                ])
                model3.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
                model3.fit(X_train, y_train, epochs = 0)
                scores.append(model3.evaluate(X_test, y_test))
# подгоняем данные под 6 входных параметров, по циклам меняем кол-во нейронов в скрытых слоях, собираем сеть, считаем accuracy
max(scores) # поверьте мне accuracy: 0.2474 не меняется от кол-ва нейронов
```
