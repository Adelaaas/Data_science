```python
df = pd.read_csv("train (7).csv")
df
```

```python
# размерность датасета - (2000, 21)
print('Размерность датасета: ', df.shape, '\n_________________________________________\n')

# соотношение классов

df['price_range'].unique()

print('Количество объектов 0 класса: ', df[df['price_range'] == 0].shape[0])
print('Количество объектов 1 класса: ', df[df['price_range'] == 1].shape[0])
print('Количество объектов 2 класса: ', df[df['price_range'] == 2].shape[0])
print('Количество объектов 3 класса: ', df[df['price_range'] == 3].shape[0])
```

*   Классификация - определение принадлежности объекта к какой-либо известной группе
*   Ансамбль - использование нескольких моделей для решения одной задачи с целью повышения произодительности
*   Нейронная сеть - один из самых распространённых методов машинного обучения, в котором используются "нейроны", работающие с входными данными
*   Метрики классификации : accuracy, precision, recall, F2 score


```python
X = df.drop(['price_range'], axis=1)
y = df['price_range']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)
```

```python
# KNN

knn = KNeighborsClassifier()

knn.fit(X_train, y_train)

pred_knn = knn.predict(X_test)

print('Accuracy KNN: ', accuracy_score(y_test, pred_knn))
```

```python
# CATBOOSTCLASSIFIER

catboost = CatBoostClassifier(depth=10)

catboost.fit(X_train, y_train)

pred_catboost = catboost.predict(X_test)

print('\nAccuracy CatBoostClassifier: ', accuracy_score(y_test, pred_catboost))
```

```python
# НЕЙРОННАЯ СЕТЬ
from keras.utils import to_categorical
y = to_categorical(y, num_classes=4)

X = df.drop(['blue', 'fc', 'touch_screen', 'mobile_wt', 'clock_speed', 'price_range'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

model = Sequential()

model.add(Dense(10, input_dim=15, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer="adam", metrics=['accuracy', 'Recall'])

model.fit(X_train, y_train, epochs=50)
```
