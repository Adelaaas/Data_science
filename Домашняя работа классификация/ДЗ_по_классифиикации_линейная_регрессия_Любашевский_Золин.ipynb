```python
#логистическая регрессия со стандартизированными данными
X_train = APStrain.drop(['satisfaction_neutral or dissatisfied', 'satisfaction_satisfied'], axis = 1)
y_train = APStrain['satisfaction_satisfied']
X_test = APStest.drop(['satisfaction_neutral or dissatisfied', 'satisfaction_satisfied'], axis = 1)
y_test = APStest['satisfaction_satisfied']

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_st_train = scaler.fit_transform(X_train)
X_st_test = scaler.fit_transform(X_test)

model = LogisticRegression()
model.fit(X_st_train, y_train)
answers_pred = model.predict(X_st_train)
print(model.score(X_st_test, y_test))
```
