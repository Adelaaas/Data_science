```python
df = pd.read_csv('train (7).csv')
df
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['price_range'], axis=1)
Y = df['price_range']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
#логистическая регрессия 
from sklearn.linear_model import LogisticRegression 
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
answer_pred = model.predict(X_train)
print(model.score(X_test, y_test))
```

```python
from catboost import CatBoostRegressor
model = CatBoostRegressor()
model.fit(X_train, y_train)
answer_pred = model.predict(X_train)
print(model.score(X_test, y_test))
```

```python
#нейронная сеть
from keras.utils import np_utils
Y = np_utils.to_categorical(Y, num_classes=5)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()

#входной слой 
model.add(Dense(5, input_dim=20, activation='relu'))

#скрытые слои
model.add(Dense(5, activation='relu'))
model.add(Dense(9, activation='relu'))
model.add(Dense(13, activation='relu'))
model.add(Dense(18, activation='relu'))

#выходной слой
model.add(Dense(5, activation='softmax'))
```

```python
model.compile(optimizer="rmsprop", loss="categorical_crossentropy", metrics=['accuracy'])
 
model.fit(X, Y, epochs = 30)
```

```python
#в итоге лучше всего себя показала ансамблевая модель catboost (с эффективностью 93%)
#чуть хуже нейронная сеть (с эффективностью 88%), например на 27 эпохе 
#хуже всего себя показала логистическая регрессия (с эффективностью 62%)
```
