```
import pandas as pd
df = pd.read_csv("country_vaccinations.csv")
df
```

```
#размерномть датафрейма df
print(df.shape)
```

```
#какие типы данных представлены в наборе
print(df.dtypes)
```

```
#сколько пропущеных значений 
print(df.isna().sum())
```

```
#какие столбцы имеют категориальный признак

for i in range(len(df.dtypes)):
  if df.dtypes[i] == 'object' :
    print(df.columns[i], df.dtypes[i])
```

```
df = df.drop(['people_fully_vaccinated', 'people_fully_vaccinated_per_hundred', 'daily_vaccinations_raw'], axis = 1)
```

```
df
```

```
for i in range(len(df.dtypes)):
  if df.dtypes[i] == 'float64' :
    print(df.columns[i], df.dtypes[i])
```

```
df = df.fillna(df.mean()) #замена пустых значений средними
df
```

```
print(df.isna().sum())
```
